```python
from transformers import pipeline
from langchain_huggingface import HuggingFacePipeline
from langchain.tools import tool
from langchain_core.messages import HumanMessage
import requests
```

```python
pip install -U langchain langchain-ollama ollama
```

```python
from langchain_ollama import ChatOllama
```

```python
pipe= pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    max_new_tokens=500,
    temperature=0.7,
    return_full_text=False
)
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

```python

#@tools
@tool
def get_multiplication_table(number: int) -> str:
    """Returns the multiplication table for a given number."""
    table = ""
    for i in range(1, 11):
        table += f"{number} x {i} = {number * i}\n"
    return table
```

```python
print(get_multiplication_table.invoke({
    "number": 5
}))
```

```python
from langgraph.prebuilt import create_react_agent
```

```python
from langchain_huggingface import ChatHuggingFace
```

```python
chat_model = ChatHuggingFace(llm=llm)
```

```python
pip install langchainhub
```

```python
from langchainhub.client import Client
```

```python
hub_client = Client()
react_prompt = hub_client.pull("hwchase17/react")
```

```python
agent = create_react_agent(
    model=chat_model,
    tools=[get_multiplication_table],
    prompt=react_prompt
)
```

```python
result = agent.invoke(
   {
       "messages":[
              HumanMessage(content= "multiplication table of 7?")
         ]
   }
)
```

```python
# Use the langgraph agent directly instead of AgentExecutor
response = agent.invoke({
    "messages": [
        HumanMessage(content="What is the multiplication table for 7?")
    ]
})
```
